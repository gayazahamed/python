```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
df = pd.DataFrame(data)
plt.scatter(df.duration, df.calories,marker='*',color='red', s=30)
plt.xlabel('grade')
plt.ylabel('Count')
plt.title("titleee")
plt.show()
```
